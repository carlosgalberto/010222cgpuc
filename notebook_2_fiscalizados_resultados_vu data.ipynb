Dataset fiscalizados_resultados_vu

Junção do dataset dos fiscalizados com resultado (notebook_1) e as informações de histórico de viagens, renda, ocorrências, tributação, retenção etc.
Os dados foram obtidos por meio de um script que acessa as informações via requisições web e leitura de json.

```python
import pandas as pd
import numpy as np
pd.set_option('display.max_columns',None)
```

```python
vu1 = pd.read_csv('VU1.csv',sep=';',encoding = 'latin1',dtype={'vu_cod':'category'})
vu2 = pd.read_csv('VU2.csv',sep=';',encoding = 'latin1',dtype={'vu_cod':'category'})
vu3 = pd.read_csv('VU3.csv',sep=';',encoding = 'latin1',dtype={'vu_cod':'category'})
vu4 = pd.read_csv('VU4.csv',sep=';',encoding = 'latin1',dtype={'vu_cod':'category'})
vu5 = pd.read_csv('VU5.csv',sep=';',encoding = 'latin1',dtype={'vu_cod':'category'})
frames = [vu1,vu2,vu3,vu4,vu5]
df_vu = pd.concat(frames)
df_vu = df_vu.drop_duplicates()
df_vu.info()
```

-------------------------------------------
Dataset dos fiscalizados com resultado para merge (notebook_1)

```python
df_fiscalizados_resultado = pd.read_csv('fiscalizados_resultado.csv',sep=';',dtype={'Código do VU':'category'})
# df_fiscalizados_resultado
```

-----------------------------------------------
Merge

```python
df_fisc_resul_vu = pd.merge(df_fiscalizados_resultado,df_vu, left_on='Código do VU',right_on='vu_cod', how='left')
df_fisc_resul_vu.info()
```
